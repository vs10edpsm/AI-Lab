```python
def dfs_water_jug():
    # capacities of the jugs
    CAP_X = 4   # 4-gallon jug
    CAP_Y = 3   # 3-gallon jug

    start = (0, 0)
    goal = (2, 0)

    stack = [(start, [start])]   # stack stores (current_state, path)
    visited = set()              # to avoid revisiting states

    while stack:
        (x, y), path = stack.pop()

        if (x, y) in visited:
            continue

        visited.add((x, y))

        if (x, y) == goal:
            return path

        next_states = []

        # Fill operations
        next_states.append((CAP_X, y))   # fill 4-gallon jug
        next_states.append((x, CAP_Y))   # fill 3-gallon jug

        # Empty operations
        next_states.append((0, y))       # empty 4-gallon jug
        next_states.append((x, 0))       # empty 3-gallon jug

        # Pour 4-gallon -> 3-gallon
        transfer = min(x, CAP_Y - y)
        next_states.append((x - transfer, y + transfer))

        # Pour 3-gallon -> 4-gallon
        transfer = min(y, CAP_X - x)
        next_states.append((x + transfer, y - transfer))

        for state in next_states:
            if state not in visited:
                stack.append((state, path + [state]))

    return None
```

```python
dfs_water_jug()
```
