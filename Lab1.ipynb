```python
def is_prime(g):
    calc = 0
    for i in range(2,g):
        if (g % i) == 0:
            calc = calc + 1
    return calc
```

```python
i = 0
j = 2
while(i < 100):
    if(is_prime(j) == 0):
        print(j)
        i = i + 1
    j = j + 1
```
