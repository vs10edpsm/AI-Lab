```python
import random

distance_matrix = [
    [0, 3, 2],
    [3, 0, 1],
    [2, 1, 0]
]

cities = ['A', 'B', 'C']
```

```python
# Generate neighbors using swap
def get_neighbors(tour):
    neighbors = []
    n = len(tour)
    for i in range(n):
        for j in range(i + 1, n):
            neighbor = tour[:]
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            neighbors.append(neighbor)
    return neighbors
```
