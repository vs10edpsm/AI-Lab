```python
def isPrime(num):
    sum = 0
    for i in range(2, int(math.sqrt(num) + 1)):
        if (num % i) == 0:
            sum = sum + 1
    if(sum > 0):
        return False
    else:
        return True

for i in range(2,51):
    if isPrime(i):
        print(i, end=' ')
```
